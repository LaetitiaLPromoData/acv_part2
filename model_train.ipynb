```python
data = pd.read_csv('database_2.csv')
data.drop("Unnamed: 0", axis=1, inplace=True)
data
```

```python

X , y = data.drop(columns=['target']), data['target']
```

```python
import pickle
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


# Création de la pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),             # Étape 1 : Standardisation
    ('knn', KNeighborsClassifier(n_neighbors=5))  # Étape 2 : Modèle KNN
])

# Entraînement de la pipeline
pipeline.fit(X, y)



# Sauvegarde de la pipeline
with open("knn_pipeline_2.pkl", "wb") as f:
    pickle.dump(pipeline, f)

print("Pipeline sauvegardée sous 'knn_pipeline.pkl'")
```
